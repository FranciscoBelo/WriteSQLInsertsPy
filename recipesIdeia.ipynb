```python
import random

# Dictionary of healthy recipes with preparation instructions
recipes = {
    "Grilled Vegetable Quinoa Salad": [
        "Cook quinoa according to package instructions.",
        "Grill your choice of vegetables (suggestions: zucchini, bell pepper, onion) and cut them into pieces.",
        "In a bowl, mix the cooked quinoa with the grilled vegetables.",
        "Season with olive oil, lemon juice, salt, and pepper to taste.",
        "Serve cold."
    ],
    "Steamed Chicken Breast with Vegetables": [
        "Season chicken breasts with salt, pepper, garlic, and lemon juice.",
        "Steam the vegetables until tender (suggestions: broccoli, carrot, green beans).",
        "Grill the chicken breasts in a skillet until golden brown.",
        "Serve the grilled chicken with steamed vegetables."
    ],
    "Turkey Wrap with Lettuce and Tomato": [
        "Heat a skillet and grill the turkey slices until crispy.",
        "Spread a thin layer of mayonnaise on a whole wheat tortilla.",
        "Place the grilled turkey slices, lettuce, and tomato on the tortilla.",
        "Roll up the wrap and cut it in half before serving."
    ],
    "Baked Salmon with Broccoli": [
        "Season the salmon fillet with salt, pepper, lemon juice, and garlic.",
        "Bake the salmon in the oven at 200°C for about 15 minutes.",
        "Steam the broccoli until tender.",
        "Serve the baked salmon with the steamed broccoli."
    ],
    "Egg White Omelette with Spinach": [
        "Beat egg whites in a bowl and season with salt and pepper.",
        "Heat a non-stick skillet and add the beaten egg whites.",
        "Once the egg whites start to set, add chopped spinach.",
        "Fold the omelette in half and cook for a few more minutes.",
        "Serve hot."
    ],
    "Scrambled Tofu with Mushrooms": [
        "Heat some oil in a skillet.",
        "Add crumbled tofu and sliced mushrooms.",
        "Season with salt, pepper, onion powder, and garlic powder.",
        "Cook until the tofu is lightly golden and the mushrooms are tender.",
        "Serve hot."
    ],
    "Chickpea Salad with Tomato and Cucumber": [
        "Drain and rinse chickpeas.",
        "Dice tomatoes and cucumbers.",
        "In a bowl, mix the chickpeas, tomatoes, and cucumbers.",
        "Season with olive oil, balsamic vinegar, salt, and pepper to taste.",
        "Finish with fresh basil leaves.",
        "Serve cold."
    ],
    "Zucchini Noodles with Homemade Tomato Sauce": [
        "Cut the zucchini into noodle shapes using a spiralizer or grater.",
        "In a saucepan, heat olive oil and sauté minced garlic and diced onions.",
        "Add chopped tomatoes and cook on low heat until it forms a sauce.",
        "Season the sauce with salt, pepper, and your preferred herbs.",
        "In a skillet, sauté the zucchini noodles for a few minutes.",
        "Serve the zucchini noodles with the homemade tomato sauce."
    ],
    "Fruit Smoothie Bowl with Granola": [
        "In a blender, add your choice of fruits (suggestions: banana, strawberry, mango).",
        "Add natural yogurt and a splash of water or plant-based milk.",
        "Blend until smooth and creamy.",
        "In a bowl, pour the smoothie and top with granola and fresh fruits.",
        "Serve chilled."
    ],
    "Grilled Chicken Salad with Mixed Greens": [
        "Season chicken fillets with salt, pepper, lemon juice, and your preferred herbs.",
        "Grill the chicken fillets until they are golden brown and fully cooked.",
        "In a bowl, mix the greens (suggestions: lettuce, arugula, watercress).",
        "Add halved cherry tomatoes and grated carrots.",
        "Slice the grilled chicken and place it on top of the greens.",
        "Season with olive oil and balsamic vinegar to taste.",
        "Serve cold."
    ]
}
# Dictionary to map weekdays to recipe indices
weekdays = {
    1: "Monday",
    2: "Tuesday",
    3: "Wednesday",
    4: "Thursday",
    5: "Friday"
}

# Function to generate the meal plan
def generate_meal_plan(recipes, weekdays):
    meal_plan = {}
    days = list(weekdays.keys())

    for day in days:
        # Select a random recipe for the weekday
        recipe = random.choice(list(recipes.keys()))
        # Add the day and recipe to the meal plan
        meal_plan[weekdays[day]] = {"Recipe": recipe, "Steps": recipes[recipe]}
        # Remove the selected recipe from the list to avoid repetition
        del recipes[recipe]

    return meal_plan

# Generate the meal plan for the week
weekly_plan = generate_meal_plan(recipes, weekdays)

# Print the meal plan
for day, meal in weekly_plan.items():
    print(f"{day}: {meal['Recipe']}")
    print("Steps:")
    for i, step in enumerate(meal['Steps'], start=1):
        print(f"{i}. {step}")
    print()
```

```python
import requests
import pandas as pd

# API endpoint and your API key
api_endpoint = "https://api.spoonacular.com/recipes/random"
api_key = "YOUR_API_KEY"

# Parameters for the API request
params = {
    "number": 50,  # Number of random recipes to fetch
    "apiKey": api_key
}

# Send the API request
response = requests.get(api_endpoint, params=params)
data = response.json()

# Extract the recipe titles from the API response
recipes = [recipe["title"] for recipe in data["recipes"]]

# Create a DataFrame with the recipes
df = pd.DataFrame(recipes, columns=["Recipes"])

# Add an index column
df.index = range(1, len(df) + 1)

# Print the table
print(df)
```
